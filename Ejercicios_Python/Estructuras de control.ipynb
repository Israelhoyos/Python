```python
num = int(input("Ingrese un número: "))

if num % 2 == 0:
    print(num, "es un número par")
else:
    print(num, "es un número impar")
```

```python
x = 5

if x > 10:
    print("x es mayor que 10")
elif x > 5:
    print("x es mayor que 5 pero menor o igual que 10")
elif x > 2:
    print("x es mayor que 2 pero menor o igual que 5")
```

Es importante considerar que, para verificar una variable booleana, no es necesario utilizar un operador de comparación, sino que basta con poner la variable.

```python
# El resultado de una comparación, siempre devolverá un valor de tipo booleano (True o False).
variable_booleana = 1 == 1

if variable_booleana:
    print(variable_booleana)
```

## <font color='green'>Actividad 1</font>

Supongamos que estamos escribiendo un programa para una empresa de mensajería que se encarga de enviar paquetes a diferentes destinos. En este contexto, la variable **"x"** representa el código postal de un destino al que se está enviando un paquete.

En nuestra empresa de logística, aseguramos que todos los paquetes 🎁 lleguen a su destino 🏙️. Si el paquete va a la ciudad A, B o C, lo entregamos allí. ¡👌 Siempre cuidando de que lleguen en óptimas condiciones! Pero, si el destino es otro 🤷‍♀️, lamentablemente no podemos entregarlo ahí 🚫.

Escribe un programa que represente esta situación.

### <font color='green'>Solución</font>

```python
x = 'a'

if x == 'a':
    print("📦🚚 Si el paquete va a la ciudad A 🏙️, lo entregamos allí. 🎁")

elif x == 'b':
    print("📦🚚 Si el paquete va a la ciudad B 🏙️, lo entregamos allí. 🎁")

elif x == 'c':
    print("📦🚚 Si el paquete va a la ciudad C 🏙️, lo entregamos allí. 🎁")

else:
    print("📦🤷‍♀️ Si el paquete va a otro destino, lo sentimos, pero 🙅‍♀️ no podemos entregarlo ahí. 🚫")
```

## <font color='green'>Actividad 2</font>

Supongamos que una tienda en línea que ofrece envío a domicilio basado en una tabla de descuentos según el monto del pedido.

- Si el pedido es mayor o igual a 1500, se realiza un envío express gratis.
- Si es mayor o igual a 1000 pero menor a 1500, el envío es gratis pero se realiza mediante el servicio estándar;.
- Si es mayor o igual a 500 pero menor a 1000, el envío tiene un costo de 10 y se realiza mediante el servicio estándar.
- Si el pedido es menor a 500, no se puede procesar.

### <font color='green'>Solución</font>

```python
# Puedes modificar este valor. ¡O podría solicitarse con input!
monto_pedido = 1000
```

```python
if monto_pedido >= 1500:
    print("🎉 Tu pedido califica para envío express gratis 🚚")

elif monto_pedido >= 1000 and monto_pedido < 1500:
    print("🎉 Tu pedido califica para envío estándar gratis 🚛")

elif monto_pedido >= 500 and monto_pedido < 1000:
    print("🚛 Tu pedido será enviado por envío estándar por un costo de $10")

else:
    print("🙁 Lo sentimos, no podemos procesar tu pedido ya que el monto mínimo de pedido es de $500")
```
